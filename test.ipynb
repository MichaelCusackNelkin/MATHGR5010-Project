```python
import databento as db
import numpy as np
import pandas as pd
import glob
```

```python
sym_1 = "AAPL"

# Get all CSV files in the data folder per desired symbol
sym1_files = glob.glob(f'data/EQUS-OHLCV-1d/equs-mini-20230328-20251003.ohlcv-1d.{sym_1}.dbn.zst')

# Read all files into a list of dataframes
dfs_sym1 = [db.DBNStore.from_file(file).to_df() for file in sym1_files] #UTC time

# Concatenate and sort
sym1 = pd.concat(dfs_sym1, join='inner').sort_index()

# Combine into one dataframe
data = pd.DataFrame({"sym1": sym1['close']}).ffill() #forward fill any missing times

data
```
