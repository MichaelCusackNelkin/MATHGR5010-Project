```python
import yfinance as yf
import pandas as pd

# manually copied tickers from top 100 most active from Yfinance :(
tickers = [
    "ONDS", "OPEN", "NVDA", "TSLA", "BBAI", "SOFI", "PLUG", "PFE", "QUBT", "DNN",
    "ACHR", "F", "INTC", "RIVN", "BMNR", "AMZN", "MARA", "NIO", "KVUE", "RIG",
    "PLTR", "SNAP", "GOOGL", "HPE", "LNW", "AAL", "RGTI", "GRAB", "WULF", "CIFR",
    "HOOD", "AMD", "XPEV", "NOK", "IREN", "AAPL", "NU", "GOOG", "BBD", "WBD",
    "BTG", "BTE", "T", "QBTS", "SOUN", "HIMS", "SMCI", "MU", "MTSR", "OWL",
    "CLSK", "B", "BAC", "CRWV", "ABEV", "VALE", "RXRX", "EOSE", "NBIS", "APLD",
    "RKT", "SMR", "JOBY", "CRCL", "NFLX", "CMCSA", "HBAN", "IONQ", "QS", "CLF",
    "COMP", "CSCO", "BULL", "RIOT", "GGB", "CMG", "YMM", "PCG", "VZ", "KEY",
    "AGNC", "HPQ", "PYPL", "MSTR", "CZR", "RKLB", "AMCR", "NGD", "BABA", "LCID",
    "PR", "AUR", "LYFT", "PATH", "DELL", "COIN", "JD", "META", "STLA"
]

rows = []
for t in tickers:
    info = yf.Ticker(t).info
    vol = info.get("regularMarketVolume", None)
    rows.append({"Ticker": t, "Volume": vol})

df = pd.DataFrame(rows)
df_sorted = df.sort_values("Volume", ascending=False)
top50 = df_sorted.head(50)
top50_list = top50["Ticker"].tolist()

print(top50)
```

```python
hist = yf.download(top50_list, start="2020-01-01", end="2025-01-01")
prices = hist["Close"].copy()

# Drop tickers that are all NaN 
prices = prices.dropna(axis=1, how="all")
prices = prices.ffill().bfill()

# daily returns
returns = prices.pct_change().dropna()
returns
```
